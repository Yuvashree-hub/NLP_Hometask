```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk import pos_tag, FreqDist
from nltk.util import ngrams
from textblob import TextBlob

# Download required resources (Run once)
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')

text = "Python is widely used for Natural Language Processing."

# 1. Tokenization
print("Word Tokens:")
tokens = word_tokenize(text)
print(tokens)

print("\nSentence Tokens:")
print(sent_tokenize(text))

# 2. Stop Words
stop_words = set(stopwords.words("english"))
filtered = [word for word in tokens if word.lower() not in stop_words]
print("\nAfter Removing Stop Words:")
print(filtered)

# 3. Stemming
stemmer = PorterStemmer()
print("\nStemming:")
print(stemmer.stem("running"))

# 4. Lemmatization
lemmatizer = WordNetLemmatizer()
print("\nLemmatization:")
print(lemmatizer.lemmatize("running", pos="v"))

# 5. POS Tagging
print("\nPOS Tagging:")
print(pos_tag(tokens))

# 6. Frequency Distribution
fd = FreqDist(tokens)
print("\nFrequency of 'Python':")
print(fd["Python"])

# 7. N-grams
print("\nBigrams:")
print(list(ngrams(tokens, 2)))

# 8. Sentiment Analysis
blob = TextBlob("I love Python and NLP")
print("\nSentiment:")
print(blob.sentiment)

# 9. String Operations
print("\nLowercase:")
print(text.lower())

print("\nReplace:")
print(text.replace("Python", "AI"))
```
