```python
import nltk
import string
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import FreqDist

# Download resources (Run once)
nltk.download('punkt')
nltk.download('stopwords')

# News Articles
news = [
    "Artificial intelligence is transforming healthcare and education systems",
    "AI and machine learning are the future of technology and innovation",
    "Healthcare systems are improving with artificial intelligence solutions",
    "Education technology is growing rapidly with AI based tools"
]

# 1. Convert to lowercase
text = " ".join(news).lower()

# 2 & 3. Remove stopwords, punctuation and tokenize
stop_words = set(stopwords.words("english"))
tokens = word_tokenize(text)

words = []
for word in tokens:
    if word not in stop_words and word not in string.punctuation:
        words.append(word)

print("Tokens:")
print(words)

# 4. Word Frequency
fd = FreqDist(words)

print("\nWord Frequency:")
print(fd)

# 5. Vocabulary
vocabulary = set(words)
print("\nVocabulary:")
print(vocabulary)

# 6. Top 5 Frequent Words
print("\nTop 5 Frequent Words:")
print(fd.most_common(5))

# 7. Type Token Ratio
ttr = len(vocabulary) / len(words)
print("\nType Token Ratio:", round(ttr,2))

# 8. Dominant Topic
topics = ["ai","artificial","intelligence","technology","healthcare","education"]

topic_freq = {}

for topic in topics:
    topic_freq[topic] = fd[topic]

dominant = max(topic_freq, key=topic_freq.get)

print("\nDominant Topic:", dominant)

# 9. Bar Chart
top = fd.most_common(5)

x = [i[0] for i in top]
y = [i[1] for i in top]

plt.bar(x, y)
plt.title("Top 5 Frequent Words")
plt.xlabel("Words")
plt.ylabel("Frequency")
plt.show()
```
