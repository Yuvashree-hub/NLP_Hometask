```python
import numpy as np

# -----------------------------
# POS Tags and Vocabulary
# -----------------------------
tags = ["NOUN", "VERB", "DET"]
words = ["the", "dog", "barks"]

tag_to_idx = {tag: i for i, tag in enumerate(tags)}
word_to_idx = {word: i for i, word in enumerate(words)}

# -----------------------------
# Log Probabilities
# -----------------------------
start_prob = np.log(np.array([0.3, 0.2, 0.5]))

transition = np.log(np.array([
    [0.2, 0.6, 0.2],   # NOUN
    [0.5, 0.2, 0.3],   # VERB
    [0.8, 0.1, 0.1]    # DET
]))

emission = np.log(np.array([
    [0.1, 0.8, 0.1],   # NOUN
    [0.1, 0.1, 0.8],   # VERB
    [0.9, 0.05, 0.05]  # DET
]))

# -----------------------------
# Input Sentence
# -----------------------------
sentence = ["the", "dog", "barks"]
obs = [word_to_idx[word] for word in sentence]

n_tags = len(tags)
T = len(sentence)

# -----------------------------
# Viterbi Tables
# -----------------------------
viterbi = np.zeros((n_tags, T))
backpointer = np.zeros((n_tags, T), dtype=int)

# Initialization
viterbi[:, 0] = start_prob + emission[:, obs[0]]

# Viterbi Algorithm
for t in range(1, T):
    scores = viterbi[:, t-1][:, None] + transition
    backpointer[:, t] = np.argmax(scores, axis=0)
    viterbi[:, t] = np.max(scores, axis=0) + emission[:, obs[t]]

# Backtracking
best_path = np.zeros(T, dtype=int)
best_path[-1] = np.argmax(viterbi[:, -1])

for t in range(T - 2, -1, -1):
    best_path[t] = backpointer[best_path[t + 1], t + 1]

predicted_tags = [tags[i] for i in best_path]

# -----------------------------
# Evaluation
# -----------------------------
true_tags = ["DET", "NOUN", "VERB"]

correct = sum(p == t for p, t in zip(predicted_tags, true_tags))
accuracy = (correct / len(true_tags)) * 100

# -----------------------------
# Output
# -----------------------------
print("Sentence       :", sentence)
print("Predicted Tags :", predicted_tags)
print("True Tags      :", true_tags)
print("Accuracy       : {:.2f}%".format(accuracy))
```
