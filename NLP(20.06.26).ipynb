```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, RegexpParser, FreqDist
from nltk.util import ngrams
from nltk.classify import NaiveBayesClassifier
from nltk.metrics import accuracy
from nltk.sem import Expression

# Download resources (Run once)
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

news = """
Microsoft announced a new AI platform in Chennai.
The company expects the technology to improve business productivity.
"""

# Parsing
tokens = word_tokenize(news)
tags = pos_tag(tokens)

grammar = "NP: {<DT>?<JJ>*<NN.*>+}"
parser = RegexpParser(grammar)
tree = parser.parse(tags)

print("Parsing:")
print(tree)

# Named Entity Recognition
ner = nltk.ne_chunk(tags)

print("\nNamed Entities:")
print(ner)

# Classification
train_data = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"cricket": True, "match": True}, "Sports"),
    ({"government": True, "election": True}, "Politics"),
    ({"business": True, "market": True}, "Business")
]

classifier = NaiveBayesClassifier.train(train_data)

features = {
    "AI": "AI" in news,
    "technology": "technology" in news.lower(),
    "cricket": "cricket" in news.lower(),
    "government": "government" in news.lower(),
    "business": "business" in news.lower(),
    "market": "market" in news.lower()
}

category = classifier.classify(features)

print("\nCategory:")
print(category)

# Accuracy and Precision
test = [
    ({"AI": True, "technology": True}, "Technology"),
    ({"business": True, "market": True}, "Business")
]

actual = [label for feat, label in test]
predicted = [classifier.classify(feat) for feat, label in test]

print("\nAccuracy:", accuracy(actual, predicted))

correct = sum(a == p for a, p in zip(actual, predicted))
precision = correct / len(predicted)

print("Precision:", precision)

# Frequency Distribution
words = [w.lower() for w in tokens if w.isalpha()]
fd = FreqDist(words)

print("\nWord Frequency:")
print(fd)

total = len(words)

print("\nProbability:")
for word in ["ai", "technology", "business"]:
    print(f"P({word}) = {fd[word]}/{total} = {fd[word]/total:.2f}")

# Semantics
read_expr = Expression.fromstring
expr = read_expr("Technology & Business")

print("\nSemantic Expression:")
print(expr)

# Bigrams and Trigrams
print("\nBigrams:")
print(list(ngrams(words, 2)))

print("\nTrigrams:")
print(list(ngrams(words, 3)))
```
